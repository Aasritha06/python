```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    "Transaction_Amount": [100, 200, 150, 300, 250, 400, 350, 100],
    "Account_Type": ["Savings", "Checking", "Savings", "Checking", "Savings", "Checking", "Savings", "Checking"],
    "Branch": ["A", "B", "A", "C", "B", "C", "A", "B"]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Task 1: Plot the total sum of Transaction_Amount per Account_Type
account_type_sums = df.groupby("Account_Type")["Transaction_Amount"].sum()

# Create a bar plot
account_type_sums.plot(kind="bar", color=["skyblue", "lightgreen"], figsize=(8, 5))
plt.title("Total Transaction Amount per Account Type")
plt.xlabel("Account Type")
plt.ylabel("Total Transaction Amount")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Task 2: Create a pie chart to show the percentage of transactions per Branch
branch_counts = df["Branch"].value_counts()

# Create a pie chart
branch_counts.plot(kind="pie", autopct='%1.1f%%', figsize=(7, 7), startangle=90, colors=["gold", "lightcoral", "lightblue"])
plt.title("Percentage of Transactions per Branch")
plt.ylabel("")  # Hide the y-label for aesthetics
plt.tight_layout()
plt.show()
```
